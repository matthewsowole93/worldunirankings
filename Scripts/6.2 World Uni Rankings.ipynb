```python
sub = df_times[['world_rank', 'income', 'year', 'num_students','total_score']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

```python
#### The above correlation map doesn't show ant strong correlations in the dataset. 
#### A universities total score does not necessarily mean that they'll be igh on the world ranking, as it comes in at -0.19
#### Additionally, total scoare and income have a lower correlation at 0.15 so we can assume that the universities getting more income doesnt mean so much
```

## Categorical Plots

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df_times['total_score'], bins = 20, kde = True)
```

```python
df_times.loc[df_times['total_score'] < 50, 'Uni Score'] = 'Low score'
```

```python
df_times.loc[(df_times['total_score'] >= 50) & (df_times['total_score'] < 80), 'Uni Score'] = 'Middle score'
```

```python
df_times.loc[df_times['total_score'] >= 80, 'Uni Score'] = 'High score'
```

```python
df_times['Uni Score'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="total_score", y="world_rank", hue="Uni Score", data=df_times)
```

### This shows that to be ranked high universities need to get high total scores
